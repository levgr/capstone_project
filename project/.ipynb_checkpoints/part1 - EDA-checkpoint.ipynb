### 1. Assimilate the business scenario and articulate testable hypotheses

AAVAIL management is looking for automatic system to make prediction for the next 30 days country's revenue based on the historical transaction-level data. The system is supposed to replace exisiting manual approach currently used in AAVAIL. The proposed business metric is the improvement in accuracy by the proposed AI-based service relative to the current spreadsheet-based model.
The Null hypotheses is that the proposed AI-based service accuracy will not approve the accuracy of that spreadsheet-based model. To reject the Null hypotheses we will need to show that the accuracy is improved at least at level of p-value=0.05 

### 2. State the ideal data to address the business opportunity and clarify the rationale for needing specific data.

To predict the 30 days revenue by a country, we will need to obtain the transction data per country. The ideal data will include information such as transaction's timestamp, country, price, type of service, type of user, level of utilization of the service. The ideal data will be collected for a many years without gaps or missing data.

### 4. Investigate the relationship between the relevant data, the target and the business metric.

```python
max_date = prep_df.date.max()
min_date = prep_df.date.min()
print("Total time span:",max_date  - min_date)
```

```python
total_per_country = prep_df[["country", "price"]].groupby("country").agg({"price":sum}).sort_values(by="price",ascending=False)
```

### Top 10 countries

```python
top10Countires = total_per_country.head(10)
top10Countires
```

```python
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
```

```python
top10Countires_list = list(top10Countires.index)
for ind_c,country in enumerate(top10Countires_list):
    total_price_per_month = prep_df[prep_df.country == country].set_index("date").sort_index().resample('30d').agg({"price":sum,"times_viewed":sum,"customer_id":len})
    total_price_per_month = total_price_per_month.rename(columns={"customer_id": "count"})    
    fig, axes = plt.subplots(3,1,figsize=(20,10))
    fig.autofmt_xdate()    
    for ind, column in enumerate(list(total_price_per_month.columns)):
        total_price_per_month.plot(y=column, ax = axes[ind])
        axes[ind].grid(True)
        axes[ind].title.set_text("{} - {} - {}".format(ind_c,country, column))
    
    
    
    
```

The data is evailable for less than 2 years (610 days), so this is a risk to overfit the yearly pattern. It is recommended to try to obtain data for a longer period. 
Some of the contires, e.g., Singapure have severely not regular profiles, which might indicate that some data is missing or not regulary reported. We will do some more detailed anaysis to investigate non-regularities

### 5. Articulate your findings using a deliverable with visualizations.

```python
for country in top10Countires_list:
    df = prep_df[prep_df.country == country].set_index("date")
    plt.figure()
    df.reset_index().plot.scatter(x="date",y="times_viewed",s= (500*df["price"]/df["price"].max()).clip(5),figsize = (20,10))
    plt.title(country)
    plt.grid()
```

Indeed for some countires and specially for Singapore the data for some periods is missing. In this case it is recommended to check if there was any error in data processing, communicate with the management and look for additional data sources. We will take it in account when building predication models 

It is also interesting to note lacuna in transaction at the end of the year holidays - it is very clear for contires with highest revenues, like United Kingdom

By repeating the analysis at higher resolution (one week instead of 30 days), some patterns are easier to observe. Here we again observe a local decrease in all metrics close to end of the year (both 2018 and 2019)

```python
for ind_c,country in enumerate(top10Countires_list):
    total_price_per_month = prep_df[prep_df.country == country].set_index("date").sort_index().resample('7d').agg({"price":sum,"times_viewed":sum,"customer_id":len})
    total_price_per_month = total_price_per_month.rename(columns={"customer_id": "count"})    
    fig, axes = plt.subplots(3,1,figsize=(20,10))
    fig.autofmt_xdate()    
    for ind, column in enumerate(list(total_price_per_month.columns)):
        total_price_per_month.plot(y=column, ax = axes[ind])
        axes[ind].grid(True)
        axes[ind].title.set_text("{} - {} - {}".format(ind_c,country, column))
    plt.show()
```
